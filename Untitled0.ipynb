```python
# âœ… STEP 1: Install dependencies
!pip install -q langchain openai pandas langchain-community langchain-experimental



# âœ… STEP 2: Upload Excel file
from google.colab import files
uploaded = files.upload()

# âœ… STEP 3: Load Excel into pandas DataFrame
import pandas as pd

excel_file = list(uploaded.keys())[0]
df = pd.read_excel(excel_file)
print("âœ… Excel File Loaded. Here's a preview:")
display(df.head())

# âœ… STEP 4: Setup OpenAI + LangChain Agent
from langchain.chat_models import ChatOpenAI
from langchain_experimental.agents import create_pandas_dataframe_agent
import getpass

# Get OpenAI API Key securely
openai_api_key = getpass.getpass("ğŸ”‘ Enter your OpenAI API Key: ")
# Create GPT-4 agent over the DataFrame
llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0,max_tokens=1000, openai_api_key=openai_api_key)
agent = create_pandas_dataframe_agent(llm, df, verbose=True, allow_dangerous_code=True)


# âœ… STEP 5: Ask questions in a loop
print("\nğŸ“Š Ask questions about your Excel data (type 'exit' to quit):\n")
while True:
    q = input("ğŸ§  Question: ")
    if q.lower() in ['exit', 'quit']:
        print("ğŸ‘‹ Exiting. Bye!")
        break
    try:
        response = agent.run(q)
        print("ğŸ¤– Answer:", response, "\n")
    except Exception as e:
        print("âš ï¸ Error:", e)
```
